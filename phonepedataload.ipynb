```python
import pandas as pd
import json
import os
import mysql.connector
from sqlalchemy import create_engine

```

```python
def Agg_transaction_details():
    path="D:/PHONEPAY/Data/pulse/data/aggregated/transaction/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in D['data']['transactionData']:
                    Name=z['name']
                    count=z['paymentInstruments'][0]['count']
                    amount=z['paymentInstruments'][0]['amount']
                    clm['Transaction_type'].append(Name)
                    clm['Transaction_count'].append(count)
                    clm['Transaction_amount'].append(amount)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(clm)
    return Agg_Trans
```

```python
def Agg_user_details():
    path="D:/PHONEPAY/Data/pulse/data/aggregated/user/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    clm={'State':[], 'Year':[],'Quater':[],'Registered_Users':[], 'Brand_Usage':[], 'Brand_Count':[],'Brand_Percentage':[],'Appopens':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                reguser=D['data']['aggregated']['registeredUsers']
                appopen=D['data']['aggregated']['appOpens']
                if D['data']['usersByDevice']==None:
                    clm['Brand_Usage'].append("Not_provided")
                    clm['Brand_Count'].append("Not_Provided")
                    clm['Brand_Percentage'].append("Not_Provided")
                    clm['Appopens'].append(appopen)
                    clm['Registered_Users'].append(reguser)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
                else:
                    for z in D['data']['usersByDevice']:
                        bname=z['brand']
                        bcount=z['count']
                        bpercent=z['percentage']
                        clm['Appopens'].append(appopen)
                        clm['Brand_Usage'].append(bname)
                        clm['Brand_Count'].append(bcount)
                        clm['Brand_Percentage'].append(bpercent)
                        clm['Registered_Users'].append(reguser)
                        clm['State'].append(i)
                        clm['Year'].append(j)
                        clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Agg_User=pd.DataFrame(clm)
    return Agg_User
```

```python
def Map_transaction_details():
    path="D:/PHONEPAY/Data/pulse/data/map/transaction/hover/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    clm={'District':[],'State':[], 'Year':[],'Quater':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in D['data']['hoverDataList']:
                    District=z['name']
                    count=z['metric'][0]['count']
                    amount=z['metric'][0]['amount']
                    clm['District'].append(District)
                    clm['Transaction_count'].append(count)
                    clm['Transaction_amount'].append(amount)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Map_Trans=pd.DataFrame(clm)
    return Map_Trans
```

```python
def Map_User_details():
    path="D:/PHONEPAY/Data/pulse/data/map/user/hover/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    clm={'District':[],'State':[], 'Year':[],'Quater':[], 'Registered_Users':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in D['data']['hoverData']:
                    District=z
                    count=D['data']['hoverData'][z]['registeredUsers']
                    clm['District'].append(District)
                    clm['Registered_Users'].append(count)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Map_User=pd.DataFrame(clm)
    return Map_User
```

```python
def Top_trans_details():
    path="D:/PHONEPAY/Data/pulse/data/top/transaction/country/india/state/"
    Agg_state_list=os.listdir(path)

    clm={'Pincode':[],'District':[],'State':[], 'Year':[],'Quater':[], 'District_count':[],'District_amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in range (len(D['data']['districts'])):
                    pinc=D['data']['pincodes'][z]['entityName']
                    District=D['data']['districts'][z]['entityName']
                    dcount=D['data']['districts'][z]['metric']['count']
                    damount=D['data']['districts'][z]['metric']['amount']
                    clm['Pincode'].append(pinc)
                    clm['District'].append(District)
                    clm['District_count'].append(dcount)
                    clm['District_amount'].append(damount)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Top_trans=pd.DataFrame(clm)
    return Top_trans
```

```python
def top_user_details():
    path="D:/PHONEPAY/Data/pulse/data/top/user/country/india/state/"
    Agg_state_list=os.listdir(path)

    clm={'Pincode':[],'District':[],'State':[], 'Year':[],'Quater':[], 'Registered_Users':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in range (len(D['data']['districts'])):
                    pinc=D['data']['pincodes'][z]['name']
                    District=D['data']['districts'][z]['name']
                    dregisteruser=D['data']['districts'][z]['registeredUsers']
                    clm['District'].append(District)
                    clm['Registered_Users'].append(dregisteruser)
                    clm['Pincode'].append(pinc)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Top_user=pd.DataFrame(clm)
    return Top_user
```

```python
def stateformat(x):
    for i in x:
        a=i
        a=a.title()
        if '-' in a:
            a=i.replace('-',' ')
            a=a.title()
        if 'Islands' in a:
            a=a[:-8]
        if 'District' in a:
            a=a[:-9]
        if a=='Dadra & Nagar Haveli & Daman & Diu':
            a='Dadra and Nagar Haveli and Daman and Diu'
        x.replace(to_replace=i,value=a,inplace=True)
    return x
```
